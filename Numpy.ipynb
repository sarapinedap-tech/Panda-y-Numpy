```python
## Numpy

import numpy as np

# 1. crear un arreglo

# 1.1 desde listas

vec = np.array([10, 20, 30, 40])
print("Arreglo desde listas: ")
print(vec)
print("\n")

# 1.2 desde matrices

matriz = np.array([[7, 14, 21], [28, 35, 42]])
print("Matriz: ")
print(matriz)
print("\n")

# 1.3 con rangos

rango_valores = np.arange(5, 50, 5)
print("Arreglo con rangos: ")
print(rango_valores)
print("\n")

# 1.4 arreglo con ceros y unos

ceros = np.zeros((2, 5))
unos = np.ones((4, 3))
print("Arreglo de ceros: ")
print(ceros)
print("\n")
print("Arreglo de unos: ")
print(unos)
print("\n")

# 1.5 aleatorios

aleatorio_flotante = np.random.rand(4, 4)
aleatorio_enteros = np.random.randint(50, 150, (3, 4))

print("Arreglo aleatorio: ")
print(aleatorio_flotante)
print("\n")
print("Arreglo aleatorio de enteros: ")
print(aleatorio_enteros)
```
