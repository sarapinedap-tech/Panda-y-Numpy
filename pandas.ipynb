```python
import pandas as pd ## se importa la biblioteca pandas
```

```python
df=pd.read_csv("/content/Distribución__área_de_pastos_en_el_departamento_de_Sucre_20251206.csv") ## se importa el dataframe
```

```python
df.columns = df.columns.str.lower().str.replace(" \n", "", regex=False)
df.columns = df.columns.str.replace(" ", "_", regex=False)  ##  se estandarizan los nombres de las columnas para un manejo mas facil de los datos
```

```python
df["area_en_pasto_natural_(hectarea)"] = df["area_en_pasto_natural_(hectarea)"].str.replace(",", "", regex=False).astype(float)
## se convirtió la columna a tipo numerico para poder hacer el filtro ya que antes daba error
df[df["area_en_pasto_natural_(hectarea)"] > 45000]
```

```python
df[(df["area_en_pasto_natural_(hectarea)"]) > 45000 / (df["municipios"] == "Sucre")] ## se hace un filtro para el municipio de Sucre con un area de pasto natural mayor a 45000 hectareas
```

```python
df.columns
```

```python
df["area_pasto_natural_y_mejorado"] = df["area_en_pasto_natural_(hectarea)"] + df["area_en_pasto_mejorado_(hectarea)"]
df["area_pasto_natural_y_mejorado"] ## se crea una nueva columna que corresponde a la suma de areas de pasto natural y mejorado
```

```python
df.sort_values("area_pasto_natural_y_mejorado") ## se ordenan los datos de la nueva columna creada
```
